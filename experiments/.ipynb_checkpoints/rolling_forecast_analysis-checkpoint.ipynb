# Rolling Forecast Analysis
Analysis of Kronos model rolling forecast results

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import glob
```

```python
# Load latest results file
result_files = glob.glob('results_rolling_*.json')
latest_file = max(result_files) if result_files else None

if latest_file:
    with open(latest_file, 'r') as f:
        results = json.load(f)
    print(f"Loaded {len(results)} forecast days from {latest_file}")
else:
    print("No results file found!")
```

```python
# Extract metrics
if results:
    metrics_data = []
    for day in results:
        metrics_data.append({
            'date': day['date'],
            'mae': day['metrics']['mae'],
            'mape': day['metrics']['mape'], 
            'rmse': day['metrics']['rmse']
        })
    
    metrics_df = pd.DataFrame(metrics_data)
    metrics_df['date'] = pd.to_datetime(metrics_df['date'])
    
    print("Summary Statistics:")
    print(metrics_df[['mae', 'mape', 'rmse']].describe())
```

```python
# Plot metrics over time
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(metrics_df['date'], metrics_df['mae'])
axes[0].set_title('MAE over Time')
axes[0].set_ylabel('MAE (€/MWh)')

axes[1].plot(metrics_df['date'], metrics_df['mape'])
axes[1].set_title('MAPE over Time')
axes[1].set_ylabel('MAPE (%)')

axes[2].plot(metrics_df['date'], metrics_df['rmse'])
axes[2].set_title('RMSE over Time')
axes[2].set_ylabel('RMSE (€/MWh)')

plt.tight_layout()
plt.show()
```

```python
# Sample forecast visualization (first few days)
n_samples = min(3, len(results))
fig, axes = plt.subplots(n_samples, 1, figsize=(12, 3*n_samples))
if n_samples == 1:
    axes = [axes]

for i in range(n_samples):
    day = results[i]
    hours = range(24)
    
    axes[i].plot(hours, day['actual'], 'b-o', label='Actual', markersize=4)
    axes[i].plot(hours, day['predicted'], 'r-s', label='Predicted', markersize=4)
    axes[i].set_title(f"Forecast for {day['date']} (MAE: {day['metrics']['mae']:.2f})")
    axes[i].set_xlabel('Hour')
    axes[i].set_ylabel('Price (€/MWh)')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Performance summary
if results:
    avg_mae = metrics_df['mae'].mean()
    avg_mape = metrics_df['mape'].mean()
    avg_rmse = metrics_df['rmse'].mean()
    
    print(f"\n=== ROLLING FORECAST PERFORMANCE ===")
    print(f"Forecast period: {metrics_df['date'].min().date()} to {metrics_df['date'].max().date()}")
    print(f"Number of forecast days: {len(results)}")
    print(f"Average MAE: {avg_mae:.2f} €/MWh")
    print(f"Average MAPE: {avg_mape:.2f}%")
    print(f"Average RMSE: {avg_rmse:.2f} €/MWh")
    
    best_day = metrics_df.loc[metrics_df['mae'].idxmin()]
    worst_day = metrics_df.loc[metrics_df['mae'].idxmax()]
    
    print(f"\nBest forecast: {best_day['date'].date()} (MAE: {best_day['mae']:.2f})")
    print(f"Worst forecast: {worst_day['date'].date()} (MAE: {worst_day['mae']:.2f})")
```
